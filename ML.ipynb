QUESTION 1

The total number of multiplications and additions needed to perform a matrix multiplication between (n,k) and (k,m) matrix are n*m*k, n*m*(k-1) respectively.

Explanation:
let A be the matrix with size (n,k) and B be the matrix with size (k,m).
Now, let C be the matrix multiplication of these two matrices i.e C=AB
we know that the size of the matrix C will be (n,m)

In the resultant matrix C, the element cij will be the dot product of the ith row in A and the jth column in B.

Therefore,
Each element cij requires k multiplications
To sum the k products we need k-1 additions.

we know that there are n*m elements in matrix C
Therefore,
The total number of multiplications is n*m*k
The total number of additions is n*m*(k-1)

```python
# method 1
import time
def largest(arr, k):
    arr = arr[:]
    for j in range(k):
        m = 0
        for i in range(1,len(arr)):
            if arr[i] >arr[m]:
                m=i
        if j==k-1:
            return arr[m]
        arr.pop(m)

def median(arr):
    l=len(arr)
    if l%2==1:
        return largest(arr,(l//2)+1)
    else:
        return (largest(arr,l//2)+largest(arr,(l//2)+1))/ 2

arr=[12,23,4,46,89]
starttime1=time.time()
median(arr)
endtime1=time.time()
print(endtime1-starttime1) #timetaken by the method 1
```

```python
# method 2
import time
def partition(arr,low,high):
    pivot=arr[high]
    i=low
    for j in range(low, high):
        if arr[j]<=pivot:
            arr[i],arr[j]=arr[j],arr[i]
            i+=1
    arr[i],arr[high]=arr[high],arr[i]
    return i

def quickselect(arr,low,high,k):
    if low<=high:
        a=partition(arr,low,high)
        if a==k:
            return arr[a]
        elif a<k:
            return quickselect(arr,a+1,high,k)
        else:
            return quickselect(arr,low,a-1,k)

def median(arr):
    l=len(arr)
    if l%2==1:
        return quickselect(arr,0,l-1,l//2)
    else:
        n1=quickselect(arr,0,l-1,l//2-1)
        n2=quickselect(arr,0,l-1,l//2)
        return (n1+n2)/2

arr =[12,23,4,46,89]
starttime2=time.time()
median(arr)
endtime2=time.time()
print(endtime2-starttime2) #timetaken by the method 2 (better method)
```

QUESTION 4

Given function,
Let f(x,y)= (x**2)*y+(y**3)*(sin(x))

The gradient of the following function:
with respect to x is 2*x*y+(y**3)*(cos(x)) (here y is constant)

with respect to y is (x**2)+(3*(y**2)*(sin(x))) (here x is constant)

```python
import math
def f(x,y):
    return (x**2)*y+(y**3)*(math.sin(x))
def dx(x,y):
    return 2*x*y+(y**3)*(math.cos(x))
def dy(x,y):
    return (x**2)+(3*(y**2)*(math.sin(x)))
```

```python
import jax.numpy as jnp
from jax import grad

def f(x,y):
    return ((x**2)*y)+((y**3)*jnp.sin(x))
def wrtx(x,y):
    return grad(lambda x: f(x,y))(x)
def wrty(x,y):
    return grad(lambda y: f(x,y))(y)

# Now let's take the value of x and y as 2 and 3 respectively
x=2.0
y=3.0

gradientwrtxmanu=dx(x,y)
gradientwrtymanu=dy(x,y)

gradientwrtxana=wrtx(x,y)
gradientwrtyana=wrty(x,y)

print(gradientwrtxmanu)
print(gradientwrtxana) # 0.7640352 by calculating manually

print(gradientwrtymanu)
print(gradientwrtyana) # 28.55103 by calculating manually
```
